```python
from mpl_toolkits.basemap import Basemap
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
lons = [] 
lats = []
for res in res1['results']:
    if 'decimalLatitude' in res and 'decimalLongitude' in res:
        lons.append(res['decimalLongitude'])
        lats.append(res['decimalLatitude'])
```

```python
x, y = my_map(lons, lats)
```

```python
# changed projection from 'ortho' to 'robin'
my_map = Basemap(projection='robin', lat_0=50, lon_0=-100,
              resolution='l', area_thresh=1000.0)
 
my_map.drawcoastlines()
my_map.drawcountries()
my_map.fillcontinents(color='coral')

# draw latitude and longitude
my_map.drawmeridians(np.arange(0, 360, 30))
my_map.drawparallels(np.arange(-90, 90, 30))

my_map.plot(x, y, 'bo', markersize=2)
plt.show()
```

```python
# let's zoom in, we need lower/upper left/right latitude and longitude
llcrnrlon = min(lons)
llcrnrlat = min(lats)
urcrnrlon = max(lons)
urcrnrlat = max(lats)
```

```python
# changed projection from 'ortho' to 'robin' to 'merc'

# Make this plot larger.
plt.figure(figsize=(16,12))

plt.title("Cyprinus_carpio occurence records from GBIF")

my_map = Basemap(projection='merc', lat_0=50, lon_0=-100,
              resolution='l', area_thresh=1000.0, 
                 llcrnrlon=llcrnrlon, 
                 llcrnrlat=llcrnrlat, 
                 urcrnrlon=urcrnrlon, 
                 urcrnrlat=urcrnrlat)
 
my_map.drawcoastlines()
my_map.drawcountries()
my_map.fillcontinents(color='coral')

# draw latitude and longitude
my_map.drawmeridians(np.arange(0, 360, 30))
my_map.drawparallels(np.arange(-90, 90, 30))

my_map.plot(x, y, 'bo', markersize=5)
plt.show()
```
